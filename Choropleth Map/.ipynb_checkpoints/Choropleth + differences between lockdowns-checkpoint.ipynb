```python
# load libraries
%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import geopandas as gpd
```

```python
# set the filepath and load in a shapefile
fp = "datasets/geo-data/gis-boundaries-london/ESRI/London_Borough_Excluding_MHW.shp"

map_df = gpd.read_file(fp)
```

```python
# check data type so we can see that this is not a normal dataframe, but a GEOdataframe
map_df.head()
```

```python
# now let's preview what our map looks like with no data in it
map_df.plot()
```

```python
data = pd.read_excel('../Crime/Borough Daily 2018-20.xlsx')
data2021 = pd.read_excel('../Crime/Borough Daily Data 2021.xlsx')

df = pd.concat([data, data2021])
data = df
data = data.fillna(data.mean())

data = data.rename(columns={'Date - Daily Data': 'date'})
# data = data.rename(columns={'Month-Year': 'date'})
data = data.sort_values(by="OCU Name")

data1 = data
data = data.groupby(by="OCU Name").mean()

df = data["TNO Offs"].to_frame()
df = df.reset_index()
df
```

```python
import datetime

first = datetime.datetime(2020, 3, 26)
first_end = datetime.datetime(2020, 5, 10)
second = datetime.datetime(2020, 11, 5)
second_end = datetime.datetime(2020, 12, 12)
third = datetime.datetime(2021, 1, 6)
third_end = datetime.datetime(2021, 7, 19)
```

```python
lock_list1 = []
lock_list2 = []
lock_list3 = []
before_lock_list = []

for i, row in data1.iterrows():
    date = row["date"]
    row["date"] = row["date"].strftime('%Y%m%d')
    
    if (date < first):
        before_lock_list.append(row)
    
    if (date >= first) and (date < first_end):
        lock_list1.append(row)
        
    if (date >= second) and (date < second_end):
        lock_list2.append(row)
        
    if (date >= third) and (date < third_end):
        lock_list3.append(row)
data1
```

```python
data_lock1 = pd.DataFrame(lock_list1)
data_lock2 = pd.DataFrame(lock_list2)
data_lock3 = pd.DataFrame(lock_list3)
before_lock = pd.DataFrame(before_lock_list)
before_lock
```

```python
print(data_lock1["TNO Offs"].min(), data_lock1["TNO Offs"].max())
print(data_lock2["TNO Offs"].min(), data_lock2["TNO Offs"].max())
print(data_lock3["TNO Offs"].min(), data_lock3["TNO Offs"].max())
print(before_lock["TNO Offs"].min(), before_lock["TNO Offs"].max())

print()

before_lock.drop(before_lock[before_lock['OCU Name'] == "Heathrow Airport"].index, inplace = True)
data_lock1.drop(data_lock1[data_lock1['OCU Name'] == "Heathrow Airport"].index, inplace = True)
data_lock2.drop(data_lock2[data_lock2['OCU Name'] == "Heathrow Airport"].index, inplace = True)
data_lock3.drop(data_lock3[data_lock3['OCU Name'] == "Heathrow Airport"].index, inplace = True)

data_lock1 = data_lock1.groupby(by="OCU Name").mean()
data_lock2 = data_lock2.groupby(by="OCU Name").mean()
data_lock3 = data_lock3.groupby(by="OCU Name").mean()
before_lock = before_lock.groupby(by="OCU Name").mean()


print(data_lock1["TNO Offs"].min(), data_lock1["TNO Offs"].max())
print(data_lock2["TNO Offs"].min(), data_lock2["TNO Offs"].max())
print(data_lock3["TNO Offs"].min(), data_lock3["TNO Offs"].max())
print(before_lock["TNO Offs"].min(), before_lock["TNO Offs"].max())

data_lock1.drop(data_lock1.columns.difference(['TNO Offs']), 1, inplace=True)
data_lock2.drop(data_lock2.columns.difference(['TNO Offs']), 1, inplace=True)
data_lock3.drop(data_lock3.columns.difference(['TNO Offs']), 1, inplace=True)
before_lock.drop(before_lock.columns.difference(['TNO Offs']), 1, inplace=True)


data_lock1.sort_values(by="TNO Offs", ascending=False).head()
```

```python
diff = before_lock["TNO Offs"].to_frame()
diff["first"] = data_lock1["TNO Offs"]
diff["second"] = data_lock2["TNO Offs"]
diff["third"] = data_lock3["TNO Offs"]
diff["Difference with before based on lockdown1 (%)"] = 100*(diff["TNO Offs"] - diff["first"])/diff["TNO Offs"]
diff["Difference with before based on lockdown2 (%)"] = 100*(diff["TNO Offs"] - diff["second"])/diff["TNO Offs"]
diff["Difference with before based on lockdown3 (%)"] = 100*(diff["TNO Offs"] - diff["third"])/diff["TNO Offs"]
diff.sort_values(by="Difference with before based on lockdown2 (%)").head()
```

```python
score = data_lock3.rename(columns={"TNO Offs": "Total Crime"}, index={'Barking & Dagenham': 'Barking and Dagenham', 'Hammersmith & Fulham': 'Hammersmith and Fulham', 'Kensington & Chelsea': 
                                                                     'Kensington and Chelsea'})
# score = score.sort_values(by="OCU Name",ascending=False)
score = score.groupby(by="OCU Name").sum()
score = score.reset_index()
score.sort_values(by="Total Crime", inplace=True)
score
```

```python
# join the geodataframe with the cleaned up csv dataframe
merged = map_df.set_index('NAME').join(score.set_index('OCU Name'))
merged.sort_values(by="Total Crime", inplace=True)
merged
```

```python
# set a variable that will call whatever column we want to visualise on the map
variable = 'Total Crime'

# set the range for the choropleth
vmin, vmax = merged["Total Crime"].min(), merged["Total Crime"].max()

# create figure and axes for Matplotlib
fig, ax = plt.subplots(1, figsize=(10, 6))

# create map
merged.plot(column=variable, cmap='Blues', linewidth=0.8, ax=ax, edgecolor='0.8')

# Now we can customise and add annotations

# remove the axis
ax.axis('off')

# add a title
ax.set_title('Mean Criminal Activity in London', \
              fontdict={'fontsize': '25',
                        'fontweight' : '3'})

# create an annotation for the  data source
ax.annotate('Source: MOPAC',
           xy=(0.1, .08), xycoords='figure fraction',
           horizontalalignment='left', verticalalignment='top',
           fontsize=10, color='#555555')

# Create colorbar as a legend
sm = plt.cm.ScalarMappable(cmap='Blues', norm=plt.Normalize(vmin=vmin, vmax=vmax))
sm._A = []
cbar = fig.colorbar(sm)

# this will save the figure as a high-res png. you can also save as svg
fig.savefig('Lockdown3.png', dpi=300)
```
